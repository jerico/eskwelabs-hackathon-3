# Get URLs to scrape

1. Extract all hotel urls from hotel index (e.g. https://www.tripadvisor.com.ph/Hotels-g294259-Bohol_Island_Bohol_Province_Visayas-Hotels.html). 
This page is paginated.
2. Output to flat file with all the URLs


## Bohol

```python
import requests
import time
from random import randint
from bs4 import BeautifulSoup

# Need to define base url because scraped links doesn't include it
base_url = "https://www.tripadvisor.com.ph"

# Start URL to parse
hotel_index_url = "/Hotels-g294259-Bohol_Island_Bohol_Province_Visayas-Hotels.html"
page = requests.get(base_url + hotel_index_url)
soup = BeautifulSoup(page.content, 'html.parser')

# Look for next link to parse
next_link = list(soup.find_all('a', class_="next"))

# Open file to write
f = open("bohol_hotels.txt", "a")

# While next_link has value, continue looping
while (len(next_link) > 0):
    page = requests.get(base_url + next_link[0]['href'])
    soup = BeautifulSoup(page.content, 'html.parser')
    next_link = list(soup.find_all('a', class_="next"))
    links = list(soup.find_all('a', class_="property_title"))
    for l in links:
        # Write link to file
        f.write(l['href'])
        # Write line break to file
        f.write('\n')
        # Print string that was written
        print(l['href'])
    time.sleep(randint(0, 5))
    
f.close()
```

# Download URLs to Folder

```python
# Bohol

file = open("bohol_hotels.txt", "r")

hotel_urls = file.readlines()

for u in hotel_urls:
    page = requests.get(base_url + u)
    soup = BeautifulSoup(page.content, 'html.parser')
    filename = u.split('/')[1].split('\n')[0]
    with open('bohol_hotels/' + filename, 'w') as f:
        f.write(page.text)
    print(filename)
    time.sleep(randint(0, 3))
```

```python
import csv

file = open("bohol_hotels.txt", "r")

hotel_urls = file.readlines()

with open('bohol_hotels.csv', 'a') as csvfile:
    fieldnames = [
                    'hotel_name', 'hotel_class', 'hotel_room_price', 'ta_rating', 
                    'ta_location', 'ta_cleanliness', 'ta_service', 'ta_value',
                    'review_count', 'url'
                 ]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for u in hotel_urls:
        filename = u.split('/')[1].split('\n')[0]
        page = open('bohol_hotels/' + filename, 'r')
        soup = BeautifulSoup(page, 'html.parser')
        
        hotel_name = list(soup.find(id='HEADING'))[0].string
        if (len(list(soup.find_all('div', class_="bb_price_text"))) > 0):
            hotel_room_price = list(soup.find_all('div', class_="bb_price_text"))[0].string.replace('₱', '')
        elif (len(list(soup.find_all('div', class_="chevronPriceInner"))) > 0): 
            try: 
                hotel_room_price = list(soup.find_all('div', class_="chevronPriceInner"))[0].string.replace('₱', '')
            except AttributeError:
                hotel_room_price = 0
        else:
            hotel_room_price = 0
            
        try:
            hotel_class = star_to_int(list(soup.find_all('span', class_="ui_star_rating"))[0].attrs['class'][1])
        except IndexError:
            hotel_class = 0
        
        try:
            ta_rating = list(soup.find_all('span', class_="hotels-hotel-review-about-with-photos-Reviews__overallRating--vElGA"))[0].string
        except IndexError:
            ta_rating = 0
            
        try:
            ta_location = bubble_to_int(list(soup.find_all('div', class_="hotels-hotel-review-about-with-photos-Reviews__subratings--33a7M"))[0].find_all(class_='ui_bubble_rating')[0].attrs['class'][1])
        except IndexError:
            ta_location = 0
            
        try: 
            ta_cleanliness = bubble_to_int(list(soup.find_all('div', class_="hotels-hotel-review-about-with-photos-Reviews__subratings--33a7M"))[0].find_all(class_='ui_bubble_rating')[1].attrs['class'][1])
        except IndexError: 
            ta_cleanliness = 0
            
        try:
            ta_service = bubble_to_int(list(soup.find_all('div', class_="hotels-hotel-review-about-with-photos-Reviews__subratings--33a7M"))[0].find_all(class_='ui_bubble_rating')[2].attrs['class'][1])
        except IndexError: 
            ta_service = 0
            
        try:
            ta_value = bubble_to_int(list(soup.find_all('div', class_="hotels-hotel-review-about-with-photos-Reviews__subratings--33a7M"))[0].find_all(class_='ui_bubble_rating')[3].attrs['class'][1])
        except IndexError: 
            ta_value = 0
        
        try:
            review_count = list(soup.find(class_='reviewCount'))[0].string.split(' ')[0]
        except TypeError:
            review_count = 0
            
        #no_of_rating
        #rank
        
        writer.writerow(
            {
                'hotel_name': hotel_name,
                'hotel_room_price': hotel_room_price,
                'hotel_class': hotel_class,
                'ta_rating': ta_rating,
                'ta_location': ta_location,
                'ta_cleanliness': ta_cleanliness,
                'ta_service': ta_service,
                'ta_value': ta_value,
                'review_count': review_count,
                'url': u.split('\n')[0]
            }
        )
        print(u)
        page.close()
```
